```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data, but with fake numbers
# V is uppercase (voltage), I is uppercase (current)
V = np.array([1,3,6,4,6,10,3.1])
I = np.array([.5,.3,.8,.4,.6,1,.355])
```

```python
plt.plot(I,V, 'ro') #uses matplotlib to plot voltage vs current
plt.ylabel('some numbers')
plt.xlabel('more numbers')
plt.title('title')
```

```python
m, b = np.polyfit(I,V,1) #If linear looks like it's the line that fits best
plt.plot(I, m*I + b)

plt.plot(I,V, 'ro') #uses matplotlib to plot voltage vs current
plt.ylabel('Voltage (V)')
plt.xlabel('Current (A)')
plt.title('Voltage vs Current')
print('The slope is: ',m, 'so the equation of this line is V = ',m,'*I + ',b)
#Please remember to use proper significant figures. 
```

```python
#Using the data for V and I 

fig=plt.figure(figsize=(8, 6), dpi=100)

# List of errors (could also be from list or NumPy array)
y_error = np.array([.1,.3,.6,.4,.6,.10,.1])
x_error = np.array([.1,.023,.036,.014,.06,.0310,.11])

# To use only x or y errors simple omit the relevant argument

plt.errorbar(I, V, yerr = y_error, xerr = x_error, fmt='o')
plt.plot(I, m*I + b)

# Title and axis labels
plt.ylabel('Voltage (V)')
plt.xlabel('Current (A)')
plt.title('Voltage vs Current')

plt.savefig('lab2R VvsI.png')
plt.show()
```
